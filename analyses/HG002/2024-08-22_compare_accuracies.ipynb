```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

sns.set_context("poster")
%matplotlib inline
```

```python
accuracies = {
    "tool": ["ConSTRain", "ConSTRain", "GangSTR", "GangSTR", "HipSTR", "HipSTR"],
    "filtered": ["unfiltered", "filtered", "unfiltered", "filtered", "unfiltered", "filtered"],
    "accuracy": [.9461, .9828, .9513, .9769, .9750, .9774],
    "n_loci": [1671143, 1393426, 1654569, 1372842, 1225530, 1174890],
}

n_total = 1695865
df = pd.DataFrame(accuracies).assign(frac_loci = lambda x: x["n_loci"] / n_total)
df
```

```python
fig = plt.figure(figsize=(6, 6))

ax = sns.scatterplot(
    df,
    x="frac_loci",
    y="accuracy",
    hue="tool",
    palette="colorblind",
    style="filtered",
    s=250
)
plt.grid()

_ = ax.set(
    xlabel="Fraction of loci called",
    xlim=(None, 1),
    ylabel="Accuracy",
    ylim=(.9, 1)
)

handles, labels = ax.get_legend_handles_labels()
labels[4] = ""
ax.legend(handles=handles[1:], labels=labels[1:])
_ = ax.get_legend().set(
    loc="lower left",
)
ax.get_legend().get_frame().set_alpha(1)
```
